# **1 - Boosting Theory**

The concept of **boosting** is not actually a machine learning algorithm, it is methodology *applied* to an existing machine learning algorithm, most commonly applied to the decision tree.


**Main Formula for Boosting:**

Implies that a combination of **estimators** with an applied coefficient could act as an effective **ensemble estimator**. h(x) in theory can any machine learning algorithm (estimator/learner)

Can an ensemble of **weak learners** (very simple models) be a **strong learner** when combined?

For decision tree models, we can use simple trees in place of h(x) and combine them with the coefficients on each model.

---
**1. AdaBoost (Adaptive Boosting)**

---

AdaBoost works by using an ensemble of **weak learners** and then combining them through the use of a weighted sum. 

**AdaBoost adapts by using previously created weak learners in order to adjust misclassified instances for the next created weak learner.**

**What is a weak learner?**
* A weak model is a model that is to simple to perform well on its own.
* The weakest decision tree possible would be a **stump**, one node and two leaves

  stump.JPG

Unlike a single decision tree which fit to all the data at once (fitting the data hard), AdaBoost aggregates multiple weak learners, allowing the overall **ensemble** model to **learn slowly** from the features.

AdaBoost uses an **ensemble** of **weak learners** that learn slowly in series.

Certain **weak learners** have more "say" in the final output than other due to the **multiplied** alpha parameter.

Each subsequent **t** weak learner is built using a **reweighted** data set from the **t-1** weak learner.

**Intuition of Adaptive Boosting:**
* Each stump essentially represent the strength of a feature to predict.
* Building these stumps in series and adding in the alpha parameter allows us to intelligently combine the importance of each feature together.

**Notes on Adaptive Boosting:**
* Unlike Random Forest, it is possible to overfit with AdaBoost, however it takes many tree to do this.
* Usually error has already stabilized way before enough trees are added to cause overfitting.

---
**2. Gradient Boosting**

---

**Gradient Boosting** is a very similar idea to AdaBoost, where **weak learners** are created in series in order to produce a strong **ensemble** model.

**Gradient Boosting** makes use of the **residual error** for learning.

**Gradient Boosting vs Adaboost:**
* Larger Trees allowed in Gradient Boosting.
* Learning Rate coefficient same for all weak learners.
* Gradual series learning is based on training on the **residuals** of the previous model.

**Gradient Boosting Intuition:**
* We optimize the series of trees by **learning** on the **residuals**, forcing **subsequent (selanjutnya) trees** to attempt to **correct** for the **error** in the **previous** trees.
* The trade-off is training time.
* A **learning rate** is between 0-1, which means a **very low value** would mean each **subsequent tree** has little "say", meaning **more trees** need to be created, causing a **longer** computational training time. 


**Notes on Gradient Boosting:**
* The **learning rate** is the **same** for each new model in the series, it is **not unique** to each subsequent model (unlike AdaBoost's alpha coefficient).
* Gradient Boosting is farily **robust** to **overfitting**, allowing for the number of **estimators** to be set **high** be default (~100).

## **C. AdaBoost Classifier**

### **a. Setup**

#### **Train | Test Split**

```python
X = df_clean.drop('class',axis=1)
```

```python
y = df_clean['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
X_train.shape
```

```python
X_test.shape
```

---

### **b. Create Estimator/Model**

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier(n_estimators=1)
```

---

#### **Fit/Train Model**

```python
model.fit(X_train, y_train)
```

---

#### **Feature Importances**

```python
pd.DataFrame(data=model.feature_importances_, index=X.columns, columns=['Feature Importance']).sort_values('Feature Importance')
```

```python
model.feature_importances_
```

```python
model.feature_importances_.argmax()
```

```python
X.columns[22]
```

```python
plt.figure(figsize=(8,6), dpi=75)
sns.countplot(data=df, x='odor', hue='class');
```

Odor is the most impartant feature

---

### **c. Performance Evaluation**

```python
from sklearn.metrics import classification_report, plot_confusion_matrix, accuracy_score
```

```python
model.classes_
```

```python
model.predict_proba(X_test)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
fig, ax = plt.subplots(figsize=(6,6), dpi=75)
plot_confusion_matrix(model, X_test, y_test, ax=ax, values_format='d');
```

There are still 23 misclassification (false positive) on edible mushroom. Let adjust the hyper parameters.

```python
print(classification_report(y_test, predictions))
```

At least we need to make 100% precision on edible mushroom.

---

### d. Number of Estimator vs Test Errors

```python
len(X.columns)
```

```python
test_errors = []

for n in range(1,96):
    
    model = AdaBoostClassifier(n_estimators=n)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    err = 1 - accuracy_score(y_test, preds)
    
    test_errors.append(err)
```

```python
plt.figure(figsize=(6,6), dpi=100)
plt.plot(range(1,96), test_errors);
```

After 20 estimators, the error is leveled-off
